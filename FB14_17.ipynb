```python
# äºŒè¿›åˆ¶æ˜¯ç”±ä¸€å †0å’Œ1æ„æˆçš„
# 8è¿›åˆ¶æ˜¯ç”¨0ï¼Œ1ï¼Œ2ï¼Œ3ï¼Œ4ï¼Œ5ï¼Œ6ï¼Œ7ï¼›
# 16è¿›åˆ¶æ˜¯ç”¨0ï¼Œ1ï¼Œ2ï¼Œ3ï¼Œ4ï¼Œ5ï¼Œ6ï¼Œ7ï¼Œ8ï¼Œ9ï¼Œaï¼Œbï¼Œcï¼Œdï¼Œeï¼Œfæ¥è¡¨ç¤ºã€‚
# æ¯”å¦‚è¯´ï¼Œå­—æ¯Kï¼šäºŒè¿›åˆ¶01001011ï¼Œå…«è¿›åˆ¶113ï¼Œåè¿›åˆ¶75ï¼Œåå…­è¿›åˆ¶4B
# è®¡ç®—æœºæ˜¯æœ‰è‡ªå·±çš„å·¥ä½œåŒºçš„ï¼Œè¿™ä¸ªå·¥ä½œåŒºè¢«ç§°ä¸ºâ€œå†…å­˜â€ã€‚æ•°æ®åœ¨å†…å­˜å½“ä¸­å¤„ç†æ—¶ï¼Œä½¿ç”¨çš„æ ¼å¼æ˜¯Unicodeï¼Œç»Ÿä¸€æ ‡å‡†ã€‚
# åœ¨Python3å½“ä¸­ï¼Œç¨‹åºå¤„ç†æˆ‘ä»¬è¾“å…¥çš„å­—ç¬¦ä¸²ï¼Œæ˜¯é»˜è®¤ä½¿ç”¨Unicodeç¼–ç çš„ï¼Œæ‰€ä»¥ä½ ä»€ä¹ˆè¯­è¨€éƒ½å¯ä»¥è¾“å…¥ã€‚
# ç¬¬1ï¼Œæ•°æ®åœ¨ç¡¬ç›˜ä¸Šå­˜å‚¨ï¼Œæˆ–è€…æ˜¯åœ¨ç½‘ç»œä¸Šä¼ è¾“æ—¶ï¼Œç”¨çš„æ˜¯UTF-8ï¼Œå› ä¸ºèŠ‚çœç©ºé—´ã€‚
# ç¬¬2ï¼Œä¸€äº›ä¸­æ–‡çš„æ–‡ä»¶å’Œä¸­æ–‡ç½‘ç«™ï¼Œè¿˜åœ¨ä½¿ç”¨GBKï¼Œå’ŒGB2312ã€‚
# é¢å¯¹ä¸åŒç¼–ç çš„æ•°æ®ï¼Œæˆ‘ä»¬è¦è¿›è¡Œä¸€äº›æ“ä½œæ¥å®ç°è½¬æ¢ã€‚
```

```python
# ã€encodeã€‘ï¼ˆç¼–ç ï¼‰å’Œã€decodeã€‘ï¼ˆè§£ç ï¼‰çš„ç”¨æ³•ã€‚
# å³å°†äººç±»è¯­è¨€è½¬æ¢ä¸ºè®¡ç®—æœºè¯­è¨€ï¼Œå°±æ˜¯ã€ç¼–ç ã€‘encode()ï¼›åä¹‹ï¼Œå°±æ˜¯ã€è§£ç ã€‘decode()ã€‚
# 'ä½ æƒ³ç¼–ç çš„å†…å®¹' .encode('ä½ ä½¿ç”¨çš„ç¼–ç è¡¨')
# 'ä½ æƒ³è§£ç çš„å†…å®¹' .decode('ä½ ä½¿ç”¨çš„ç¼–ç è¡¨')
print('å´æ«'.encode('utf-8'))
print('å´æ«'.encode('gbk'))
print(b'\xe5\x90\xb4\xe6\x9e\xab'.decode('utf-8'))
print(b'\xce\xe2\xb7\xe3'.decode('gbk'))
```

```python
#sample
#æŒ‰æ€»åˆ†æ’åº
file1 = open('/Users/youjia/Downloads/winnner.txt','r',encoding='utf-8') 
file_lines = file1.readlines() 
file1.close()

dict_scores = {}
list_scores = []
final_scores = []

# print(file_lines) 
# print(len('\n'))

# æ‰“å°ç»“æœä¸ºï¼š['ç½—æ©102\n', 'å“ˆåˆ©383\n', 'èµ«æ•570\n', 'é©¬å°”ç¦275\n']
# ç»è¿‡æµ‹è¯•ï¼Œå‘ç°'\n'çš„é•¿åº¦æ˜¯1ã€‚æ‰€ä»¥ï¼Œåå­—æ˜¯â€œç¬¬0ä½-å€’æ•°ç¬¬5ä½â€ï¼Œåˆ†æ•°æ˜¯â€œå€’æ•°ç¬¬4ä½-å€’æ•°ç¬¬äºŒä½â€ã€‚
# å†æ ¹æ®â€œå·¦å–å³ä¸å–â€ï¼Œå¯çŸ¥ï¼šname-[:-4],score-[-4:-1]

for i in file_lines:  # iæ˜¯å­—ç¬¦ä¸²ã€‚
#     print(i)
    name = i[:-4]  # å–å‡ºåå­—ï¼ˆæ³¨ï¼šå­—ç¬¦ä¸²å’Œåˆ—è¡¨ä¸€æ ·ï¼Œæ˜¯é€šè¿‡åç§»é‡æ¥è·å–å†…éƒ¨æ•°æ®ã€‚ï¼‰
    score = int(i[-4:-1])  # å–å‡ºæˆç»©
#     print(name)
#     print(score)
    dict_scores[score] = name  # å°†åå­—å’Œæˆç»©å¯¹åº”å­˜ä¸ºå­—å…¸çš„é”®å€¼å¯¹(æ³¨æ„ï¼šè¿™é‡Œçš„æˆç»©æ˜¯é”®)
    list_scores.append(score)

print(list_scores)
print(dict_scores)

list_scores.sort()  # reverseï¼Œé€†è¡Œï¼Œæ‰€ä»¥è¿™æ—¶åˆ—è¡¨é™åºæ’åˆ—ï¼Œåˆ†æ•°ä»é«˜åˆ°ä½ã€‚
#reverse=Falseä¸ºå‡åºæ’åºï¼›reverse=Trueä¸ºé™åºæ’åº
print(list_scores)

for i in list_scores:
    result = dict_scores[i] + str(i) + '\n'
    # print(result)
    final_scores.append(result)

print(final_scores)  # æœ€ç»ˆç»“æœ

# winner_new = open('winner_new.txt','w',encoding='utf-8') 
# winner_new.writelines(final_scores)
# winner_new.close()
```
